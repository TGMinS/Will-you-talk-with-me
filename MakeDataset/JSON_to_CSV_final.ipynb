## 대화 정보 추출 코드

```python
index=0
result_df_list=[]
for i in file_list:
    if index%100==0:  
        print("파일 넘버: ",index)
    with open(file_path+i,'r',encoding='UTF8') as file:
        data=json.load(file)
  
        for i in range(len(data['document'])):
            df=pd.DataFrame()
            talk_data=data['document'][i]
            df['talk_id']=[talk_data['id']]
            
            talk_order=0
            before_talk_id=0
            talk=""
            order_list=[]
            talk_list=[]
            
            for i2 in (talk_data['utterance']):
                if i2['speaker_id']!=before_talk_id:
                    order_list.append(talk_order)
                    talk_list.append(talk)
#                    df[f'{talk_order}']=[talk]
                    talk_order+=1
                    talk=""
                    before_talk_id=i2['speaker_id']
                    talk=talk+" "+str(i2['form'])
                else:
                    talk=talk+" "+str(i2['form'])
            df.loc[:,order_list]=[talk_list]

            result_df_list.append(df)
    index+=1
```

```python
talk_info_df=pd.concat(result_df_list[70000:74664])
talk_info_df.to_csv(f'talk_{74665}.csv',index=False,encoding="utf-8-sig")
```

```python
Person_info_df=pd.DataFrame(columns=['talk_id','topic','p1_id','p1_age','p1_occ','p1_sex','p1_birthplace','p1_residence'
                                    ,'p2_id','p2_age','p2_occ','p2_sex','p2_birthplace','p2_residence','relation','inti'])
index=0
for i in file_list:
    if index%5000 ==0:
        print("진행 index: ",index)
    # 다인의 화자가 아닌 2인 화자인 경우 아래의 작업을 수행
    if i[1]=='D':
        with open(file_path+i,'r',encoding='UTF8') as file:
            data=json.load(file)
            for i2 in range(len(data['document'])):
                talk_data=data['document'][i2]
                person_data=talk_data['metadata']['speaker']
                value_l=[talk_data['id'],talk_data['metadata']['topic'],person_data[0]['id'],
                        person_data[0]['age'],person_data[0]['occupation'],person_data[0]['sex'],
                        person_data[0]['birthplace'],person_data[0]['current_residence'],
                        person_data[1]['id'],person_data[1]['age'],person_data[1]['occupation'],
                        person_data[1]['sex'],person_data[1]['birthplace'],person_data[1]['current_residence'],
                        talk_data['metadata']['setting']['relation'],talk_data['metadata']['setting']['intimacy']]
                Person_info_df.loc[index]=value_l
                index+=1
Person_info_df
```

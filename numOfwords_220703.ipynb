```python
import pandas as pd
import numpy as np

# 데이터 출처 : AIHub, KETI R&D, '트위터 대화 시나리오' (https://aihub.or.kr/opendata/keti-data/recognition-laguage/KETI-02-008) 
df = pd.read_excel("Twitter_df.xlsx")
```

```python
df.loc[1999] = [str(df.iloc[0, 5])] * 83 # NaN값
```

```python
df.loc[1999, df.columns[0]] = df.columns[0]
df.loc[1999, df.columns[1]] = df.columns[1]
df.loc[1999, df.columns[2]] = df.columns[2]
df.loc[1999, df.columns[3]] = df.columns[3]
df.loc[1999, df.columns[4]] = df.columns[4]
df.loc[1999, df.columns[5]] = df.columns[5]
df.loc[1999, df.columns[6]] = df.columns[6]
```

```python
df.columns = [str(x) for x in range(83)]
```

```python
df
```

```python
talking_lst = [[] for _ in range(2000)]
print(len(talking_lst))
```

```python
for i in range(2000) :
    A = ''
    B = ''
    for j in range(83) :
        x = str(df.iloc[i, j])
        
        if x == str(df.iloc[0, 5]) :
            break
            
        if j % 2 == 0 :
            if len(A) == 0 :
                A += x
            else :
                A += ' ' + x
        else :
            if len(B) == 0 :
                B += x
            else :
                B += ' ' + x
            
    talking_lst[i].append(A)
    talking_lst[i].append(B)
```

# 모든 형태소에 대해 사용

## Func definition : x1 = 둘 사이 오고간 형태소의 합, x2 = 둘 사이 오고간 형태소의 차
### -log(x2/(1.1x1))으로 정의
### 분모가 1.1*x1인 이유 : 서로 대화한 합이 8번이고 서로 대화한 차이가 2번인 경우(case1)와, 서로 대화한 합이 16번이고 서로 대화한 차이가 4번인 경우(case2)를 생각했을 때, case2가 더 대화가 잘 통하는 것 같다고 생각해봤음. 이에 대한 차이를 고려하기 위함임 and 1.1이라는 계수의 조정 또는 함수의 변경 등이 필요할 것이라 예상함.

```python
from konlpy.tag import *
```

```python
komoran = Komoran()
```

```python
all_morphs = [[] for _ in range(2000)]
```

```python
for i in range(2000) :
    for j in range(2) :
        x = komoran.morphs(talking_lst[i][j])
        all_morphs[i].append(len(x))
```

```python
all_morphs_sum_diff = [[] for _ in range(2000)]

for i in range(2000) :
    x1 = all_morphs[i][0]
    x2 = all_morphs[i][1]
    all_morphs_sum_diff[i].append(x1+x2)
    all_morphs_sum_diff[i].append(abs(x1-x2))
```

```python
from math import log
all_morphs_result = []
```

```python
for i in range(2000) :
    x = all_morphs_sum_diff[i][1]/(1.1*all_morphs_sum_diff[i][0])
    a = 0
    if x == 0 :
        a = -1 * log(1/(1.1*all_morphs_sum_diff[i][0]))
    else :
        a = -1 * log(x)
    all_morphs_result.append(a)
```

```python
import matplotlib.pyplot as plt
plt.hist(all_morphs_result)
plt.show()
```

```python
tmp_range0to1 = []
tmp_range4to6 = []

for i in range(2000) :
    if all_morphs_result[i] >= 0 and all_morphs_result[i] < 1 :
        tmp_range0to1.append(i)
    elif all_morphs_result[i] >= 4 :
        tmp_range4to6.append(i)
        
print(tmp_range0to1)
print(tmp_range4to6)
```

```python
import random
tmp_range0to1_sample10 = random.sample(tmp_range0to1, 10)
tmp_range4to6_sample10 = random.sample(tmp_range4to6, 10)
```

```python
for i in tmp_range0to1_sample10 :
    print(list(df.iloc[i]))
```

```python
for i in tmp_range4to6_sample10 :
    print(list(df.iloc[i]))
```

# 명사 수만 측정

```python
only_noun = [[] for _ in range(2000)]
for i in range(2000) :
    for j in range(2) :
        a = komoran.pos(talking_lst[i][j])
        only_noun[i].append(len([x for x in a if x[1] == 'NNP' or x[1] == 'NNG']))

only_noun_sum_diff = [[] for _ in range(2000)]

for i in range(2000) :
    x1 = only_noun[i][0]
    x2 = only_noun[i][1]
    only_noun_sum_diff[i].append(x1+x2)
    only_noun_sum_diff[i].append(abs(x1-x2))
    
from math import log
only_noun_result = []
for i in range(2000) :
    if only_noun_sum_diff[i][0] == 0 :
        only_noun_result.append(0)
        continue
    x = only_noun_sum_diff[i][1]/(1.1*only_noun_sum_diff[i][0])
    a = 0
    if x == 0 :
        a = -1 * log(1/(1.1*only_noun_sum_diff[i][0]))
    else :
        a = -1 * log(x)
    only_noun_result.append(a)
import matplotlib.pyplot as plt
plt.hist(only_noun_result)
plt.show()
tmp_range0to1 = []
tmp_range2to4 = []

for i in range(2000) :
    if only_noun_result[i] >= 0 and only_noun_result[i] < 1 :
        tmp_range0to1.append(i)
    elif only_noun_result[i] >= 2 :
        tmp_range2to4.append(i)

import random
tmp_range0to1_sample10 = random.sample(tmp_range0to1, 10)
tmp_range2to4_sample10 = random.sample(tmp_range2to4, 10)
```

```python
for i in tmp_range0to1_sample10 :
    print(list(df.iloc[i]))
```

```python
for i in tmp_range2to4_sample10 :
    print(list(df.iloc[i]))
```
